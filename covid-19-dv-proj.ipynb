```python
# Visualization of top 5 state having highest Death ratio in India.
df2 = data.sort_values(by='Death Ratio', ascending=False).head()
states = df2['State/UTs']
ratio = df2['Death Ratio']
plt.barh(states, ratio)
plt.xlabel('Death Ratio (%)')
plt.ylabel('State')
plt.title('State with more death ratio in India')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
barplot = sns.barplot(data=data, x="State/UTs", y="Total Cases", hue="Death Ratio", dodge=False)
plt.title("Total Cases as per each State")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data,
    x='Active Ratio',
    y='Death Ratio',
    hue='State/UTs',  # Colors by State/UTs
    palette='viridis',  # Choose a color palette
    s=100,  # Adjust marker size
)

plt.title("Death Ratio vs Active Ratio by State/UT")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.tight_layout()  # Prevent label cutoff
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(
    data=data,
    x="State/UTs",
    y="Total Cases",
    bins=(20, 20),  # Controls granularity (nx=20, ny=20)
    cmap="viridis",  # Color map
    cbar=True,  # Show color bar (density scale)
    cbar_kws={'label': 'Density'},  # Customize color bar label
)

plt.title("Density Heatmap of Total Cases by State/UT")
plt.xticks(rotation=90)  # Rotate x-labels for readability
plt.tight_layout()  # Prevent label cutoff
```

```python


plt.figure(figsize=(12, 12))  # Increase figure size

plt.pie(
    data['Total Cases'],  # Values
    labels=data['State/UTs'],  # Labels
    autopct='%1.1f%%',  # Show percentages
    startangle=90,  # Rotate for better readability
    colors=plt.cm.tab20.colors,  # Use a nice color scheme
    wedgeprops={'edgecolor': 'white', 'linewidth': 0.5},  # Add white borders
    textprops={'fontsize': 10}  # Increase font size for readability
)

plt.title('Covid Cases (%) in All States of India', pad=20, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(12, 12))

# Use Seaborn's color palette (e.g., "husl", "pastel", "viridis")
colors = sns.color_palette("husl", len(data))

plt.pie(
    data['Discharge Ratio'],
    labels=data['State/UTs'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 0.3},
)

plt.title('Discharge Ratio (%) by State/UT in India', pad=20, fontsize=15)
```

```python
# SCATTER PLOT
plt.figure(figsize=(8,6))
plt.scatter(data['Total Cases'], data['Deaths'], color='red', alpha=0.6)
plt.title("Scatter Plot of COVID Cases vs Deaths (by State)")
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
#Histogram  Distribution of Cases
plt.figure(figsize=(8,6))
plt.hist(data['Total Cases'], bins=15, color='blue', alpha=0.7)
plt.title("Histogram of COVID-19 Cases Across States")
plt.xlabel("Total Cases")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
#Box Plot  Deaths by State
plt.figure(figsize=(8,6))
sns.boxplot(x=data['Deaths'], color="orange")
plt.title("Box Plot of COVID-19 Deaths")
plt.xlabel("Deaths")
plt.show()
```

```python
#Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    data[['Total Cases','Deaths','Discharged']].astype(float).corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Heatmap (Cases, Deaths, Discharged)")
plt.show()
```

```python
#Pairplot
numeric_data = data[['Total Cases','Deaths','Discharged']].astype(float)

sns.pairplot(numeric_data, diag_kind="kde")
plt.suptitle("Pairplot of Cases, Deaths & Discharged", y=1.02)
plt.show()
```

```python
#
plt.figure(figsize=(8,6))
sns.heatmap(
    data[['Active Ratio','Discharge Ratio','Death Ratio']].astype(float).corr(),
    annot=True,
    cmap="YlGnBu",
    linewidths=0.5
)
plt.title("Correlation Heatmap of COVID-19 Ratios")
plt.show()
```
